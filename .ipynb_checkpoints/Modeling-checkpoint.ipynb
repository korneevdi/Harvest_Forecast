```python
import pandas as pd
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import ElasticNetCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import *
from sklearn.feature_selection import mutual_info_regression, RFECV
from sklearn.model_selection import TimeSeriesSplit 
from matplotlib import pyplot as plt

import seaborn as sns

input_path = 'preprocessed_data/'
%run train_model.py
```

# Eproach
Make model for each phase to understand special conditions for each aspect corresponds to this step.

# Last phase

```python
feature_selection = FeatureSelection(input_path + 'red.csv.gz')
# importance = feature_selection.show_mutal_importance()
df = feature_selection.df.copy()
df['target'] = df['value']
df = df.drop(columns=['value'])
elastic_rigression = make_pipeline(StandardScaler(), ElasticNetCV())

models = [elastic_rigression, 
          RandomForestRegressor(n_estimators= 1000, max_depth=5, random_state=42,),
          XGBRegressor(n_estimators=600, max_depth=4, random_state=42)]
build_models(df, models, ksplit=3)
```

# XGBoost regressor  has best results

```python
features = df.drop(columns=['target', 'date'])
target = df[['target']]

train, test, train_label, test_label=train_test_split(features, target, test_size=0.33, shuffle=False)

#using XGBregression (we need to keep train data range in test)
reg = XGBRegressor(n_estimators=600, max_depth=4, random_state=42)
model = reg.fit(train, train_label)
predict = model.predict(test)
#estimate
print('rmse: ', mean_squared_error(test_label, predict, squared=False))
print(f'mape: {mean_absolute_percentage_error(test_label + 1, predict + 1)}%')


feature_importance = reg.feature_importances_
# feature_importance = model.feature_importances_
sorted_idx = np.argsort(feature_importance)
fig = plt.figure(figsize=(12, 6))
plt.barh(range(len(sorted_idx)), feature_importance[sorted_idx], align='center')
plt.yticks(range(len(sorted_idx)), np.array(features.columns)[sorted_idx])
plt.title('Feature Importance')
```

the most important features are temperature of previose phase, counts of previous phase and avg_areas turning

# White phase

```python
feature_selection = FeatureSelection(input_path + 'white.csv.gz')
# importance = feature_selection.show_mutal_importance()
# importance = feature_selection.show_mutal_importance()
df = feature_selection.df.copy()
elastic_rigression = make_pipeline(StandardScaler(), ElasticNetCV())

models = [elastic_rigression, 
          RandomForestRegressor(n_estimators= 1000, max_depth=5, random_state=42,),
          XGBRegressor(n_estimators=600, max_depth=4, random_state=42)]
build_models(df, models, ksplit=4)
```

```python
features = df.drop(columns=['target', 'date'])
target = df[['target']]

train, test, train_label, test_label=train_test_split(features, target, test_size=0.33, shuffle=False)

#using XGBregression (we need to keep train data range in test)
reg = XGBRegressor(n_estimators=600, max_depth=4, random_state=42)
model = reg.fit(train, train_label)
predict = model.predict(test)
#estimate
print('rmse: ', mean_squared_error(test_label, predict, squared=False))
print(f'mape: {mean_absolute_percentage_error(test_label + 1, predict + 1)}%')


feature_importance = reg.feature_importances_
# feature_importance = model.feature_importances_
sorted_idx = np.argsort(feature_importance)
fig = plt.figure(figsize=(12, 6))
plt.barh(range(len(sorted_idx)), feature_importance[sorted_idx], align='center')
plt.yticks(range(len(sorted_idx)), np.array(features.columns)[sorted_idx])
plt.title('Feature Importance')
```
